```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("/content/Euler.csv", error_bad_lines=False)
df2 = pd.read_csv("/content/readingsrunge.csv") 
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(df1.index,df1["w"], 'r-', linestyle='solid', linewidth=2, label='STAT1')
ax.plot(df1.index,df1["x"],'b-', linestyle='dashed', linewidth=2, label='STAT3')
ax.plot(df1.index,df1["y"],'g-', linestyle='dotted', linewidth=2, label='BCL-2')
ax.plot(df1.index,df1["z"],'y-', linestyle='dashdot', linewidth=2, label='BAX')
ax.legend()
ax.set_title("Behaviour in time using Euler method")
ax.set_xlabel("Iteration")
ax.grid()
ax.legend(loc='best')
plt.show()

```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(df2.index,df2["w"], 'r-', linestyle='solid', linewidth=2, label='STAT1')
ax.plot(df2.index,df2["x"],'b-', linestyle='dashed', linewidth=2, label='STAT3')
ax.plot(df2.index,df2["y"],'g-', linestyle='dotted', linewidth=2, label='BCL-2')
ax.plot(df2.index,df2["z"],'y-', linestyle='dashdot', linewidth=2, label='BAX')
ax.legend()
ax.set_title("Behaviour in time using Runge Kutta fourth order method")
ax.set_xlabel("Iteration")
ax.grid()
ax.legend(loc='best')
plt.show()


```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(df1.index,df1["w"], 'r-', linestyle='solid', label='Euler')
ax.plot(df2.index,df2["w"],'b-', linestyle='dashed',label='Runge Kutta')
# ax.plot(df2.index,df2["y"],'g-', label='B')
# ax.plot(df2.index,df2["z"],'y-', label='X')
ax.legend()
ax.set_title("Behaviour of STAT1 in time")
ax.set_xlabel("Iteration")
ax.set_ylabel("STAT1")
ax.grid()
ax.legend(loc='best')
plt.show()
```
